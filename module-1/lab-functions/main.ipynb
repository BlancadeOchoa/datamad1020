## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    fact=1  #el contador tiene que empezar en 1, ya que 0*n=0
    if n== 0:
        return 1
    elif n<0:
        return "Negative numbers dont have factorials"
    else:
        for n in range(1,n+1):
            fact= fact*n
        return fact
    
    
    pass
```

```python
# This will test your function 
test_factorial(factorial)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... 🧐`

```python
def mode_counter(arr):
    arr.sort()
    repetitions= []
    i=0
    while i<len(arr):
        repetitions.append(arr.count(arr[i]))
        i+= 1
    #Here we have  iterated through the list of sorted number and
    #and count the rep of each
    
    d1= dict(zip(arr, repetitions))
    
    d2={k for (k,v) in d1.items() if v == max(repetitions)}
    return d2
pass
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once.

```python
def pangram(string):
    alphabet= "abcdefghijklmnopqrstuvwxyz"
    for l in alphabet:
        if l not in string.lower():
            return False
        else:
            return True
    
    pass
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.
`Valid special characters: # @ ! $ % & ( ) ^ * [ ] { }`

```python
def check_pass(string):
    symbols = ['@','!','$','%','&','( )','#','^','*','[ ]','{ }']
    value = True
    if len(string)>8:
        value= False
    elif not any(char.islower() for char in string):
        value= False
    elif not any(char.isupper() for char in string):
        value= False
    elif not any(char.isdigit() for char in string):
        value= False
    elif not any(char in symbols for char in string):    
        value= False
    return value

pass
```

```python
# This will test your function 
test_pass(check_pass)
```
